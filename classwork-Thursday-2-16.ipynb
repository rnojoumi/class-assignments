```python
def makeBirdDict(filename):
    #this function extracts species and family as a dictionary 
    birdDict={}
    fobj = open(filename, "r", encoding = "ISO-8859-15")#open and read the file
    records= fobj.readlines()[1:] #split lines on comma
    #stick species and family into dict
    for line in records:
        species= line.split(",")[0]
        fam= line.split(",")[2]
        birdDict[species]= fam
    return birdDict
```

```python
makeBirdDict("formattednature.csv")
```
